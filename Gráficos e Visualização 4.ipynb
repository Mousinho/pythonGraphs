```python
#Gráfico de dispersão com seaborn
co_base = pd.read_csv("co2.csv")
co_base.head()
```

```python
#Gráfico de dispersão utilizando atributos conc e uptake, agrupamento pelo Type
srn.scatterplot(co_base.conc,co_base.uptake, hue = co_base.Type)
```

```python
#Seleção de registros especificos da base de dados (Quebec e Mississipi)
q = co_base.loc[co_base["Type"] == "Quebec"]
m = co_base.loc[co_base["Type"] == "Mississippi"]
```

```python
#Subgráfico (1linha 2 colunas) mostrando gráficos sobre a região
plt.figure()
plt.subplot(1,2,1)
srn.scatterplot(q.conc, q.uptake).set_title("Quebec")
plt.subplot(1,2,2)
srn.scatterplot(m.conc, m.uptake).set_title("Mississippi")
plt.tight_layout()
```

```python
#Refrigerado e não refrigerado
ch = co_base.loc[co_base["Treatment"] == "chilled"]
non = co_base.loc[co_base["Treatment"] == "nonchilled"]
```

```python
#Gráfico somente com chilled e unchilled
plt.figure()
plt.subplot(1,2,1)
srn.scatterplot(ch.conc,ch.uptake).set_title("Chilled")
plt.subplot(1,2,2)
srn.scatterplot(non.conc,non.uptake).set_title("Nonchilled")
plt.tight_layout()
```

```python
#Carregamento de outro arquivo "Cancer de esofago
esop = pd.read_csv("esoph.csv")
esop.head()
```

```python
#Gráfico entre atributos alcgp e ncontrols
srn.catplot(x = "alcgp", y = "ncontrols", data = esop, jitter = False)
```

```python
#Gráfico entre atributos alcgp e ncontrols com agrupamento
srn.catplot(x = "alcgp", y = "ncontrols", data = esop, col = "tobgp")
```

```python
#Gráficos 3d
graph3d = pd.read_csv("orchard.csv")
graph3d
```

```python
#Criando gráfico 3d, indicando atributo projection = 3d e passando três atributos (decrease, rowpos e colpos)
figura = plt.figure()
eixo = figura.add_subplot(1,1,1, projection = "3d")
eixo.scatter(graph3d.decrease, graph3d.rowpos, graph3d.colpos)
eixo.set_xlabel("decrease")
eixo.set_ylabel("rowpos")
eixo.set_zlabel("colpos")
```
